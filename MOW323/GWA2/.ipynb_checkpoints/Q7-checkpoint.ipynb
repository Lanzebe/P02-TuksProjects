```python
import numpy as np 
import importlib; 
import os
# if "add_engmod.py" in os.listdir(): print("Exists!"); import add_engmod
import structfem as sfem; importlib.reload(sfem)
import matplotlib.pyplot as plt
```

```python
node_coordinates  = np.array([[0,0],[1,0],[0,2],[1,2]]) 

print("Nodal coordinates:",node_coordinates.shape)

elem_connectivity  = np.array([[0,1],[0,2],[0,3],[1,2],[1,3],[2,3]])

print("Element connectivity:",elem_connectivity.shape)

prescribed_displacement  = np.array([[0,0,0], # Node 1, x-direction, 0 displacement
                                     [0,1,0], # Node 1, y-direction, 0 displacement
                                     [1,1,0],
                                     ])

print("Prescribed displacement:",prescribed_displacement.shape)

prescribed_force = np.array([[2,1,-9.81E6/2],[3,1,-9.81E6/2] #ode 0, y-direction, -50000 force
                            ])

print("Prescribed forces:",prescribed_force.shape)

E   = 200E9 # GPa
A   = 0.1   # m^2
rho = 7800  # kg/m^3

elem_E = np.array([E,E,E,E,E,E])           # Young's modulus of Element 0, Element 1
elem_A = np.array([A,A,A,A,A,A])           # Area of Element 0, Element 1

print("elem_E/10^9:",elem_E/1E9)
print("elem_A",elem_A)

elemprop = {"A": elem_A,
            "E": elem_E,
            "density": rho * np.ones(6)}

tmn = sfem.FEM_Truss_2D(node_coordinates,            # The coordinates of each node
                            elem_connectivity,       # The connectivity of the nodes to form elements
                            prescribed_displacement, # The prescribed displacements
                            prescribed_force,        # The prescribed loads
                            elemprop)                # The properties of the elements
```

```python
tmn.plot_system()
```

```python
tmn.solve()
```

```python
tmn.plot_system(show_element_numbers=True,
                show_node_numbers=True,
                show_displacement=True,
                magnification=1000)
```
