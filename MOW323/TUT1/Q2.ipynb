```python
import numpy as np
from numpy import sin, cos, arctan
import matplotlib.pyplot as plt

pA = [0,0]
pB = [1,0]
pC = [1,1]

F = 100
theta = np.linspace(0,2*pi,100)

gamma = arctan(pC[1]/pB[0])

def Fc():
    return F*cos(theta)/cos(gamma)

def Fb():
    return F*sin(theta)-Fc()*sin(gamma)

#print(Fc())
#print(Fb())

plt.figure()
plt.plot(theta,Fc())
plt.plot(theta,Fb(), 'r')
plt.show()
```
