# Plotting in a for loop:

```python
x = np.arange(0,10,0.001)

for idx in range(3):
    
    if idx == 0:
        y = np.sin(x)
        colour = "b-"
        label = "sin"
    elif idx == 1:
        y = np.cos(x)
        colour = "r--"
        label = "cos"        
    elif idx == 2:
        y = 0*x 
        colour = "b-."
        label = "0"        

    plt.figure(1)
    plt.plot(x,y,colour,label=label)

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
```

# Plotting in a for loop:

```python
x = np.arange(0,10,0.001)

def function1(x):
    
    return np.cos(x) - np.sin(x)

def function2(x):
    
    return np.cos(x) + np.sin(x)

def function3(x):
    
    return np.cos(x) * np.sin(x)

for idx in range(3):
    
    if idx == 0:
        y = function1(x)
        colour = "b-"
        label = "1"
    elif idx == 1:
        y = function2(x)
        colour = "r--"
        label = "2"        
    elif idx == 2:
        y = function3(x)
        colour = "b-."
        label = "3"        

    plt.figure(1)
    plt.plot(x,y,colour,label=label)

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
```
