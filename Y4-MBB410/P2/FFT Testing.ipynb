```python
import matplotlib.pyplot as plt

%matplotlib notebook
import numpy as np
from scipy.fft import fft, fftshift, fftfreq
from scipy.signal import windows as win
```

```python
fs = 2000


xs = np.arange(0, 10 + 1/(fs), 1/(fs))
print(len(xs))

hw = win.cosine(len(xs))

NumOfSines = 10


freq = np.random.uniform(low=0.1, high=50, size=(NumOfSines,))  #[1/3,2**0.5]
ampl = np.random.uniform(low=0.1, high=5, size=(NumOfSines,))

print('Frequencies:',freq)
print('Amplitudes:',ampl)

ys = np.sum(np.multiply(ampl,(np.sin(np.multiply(freq,np.transpose([2*np.pi*xs]*len(freq)))))),axis = 1)

plt.figure(1)
plt.plot(xs,ys)
plt.plot(xs,ys*hw)
plt.grid()
plt.show()

```

```python
out=fft(ys)

fv=fftfreq(len(xs),1/fs)

fig,axs=plt.subplots(2)

axs[0].plot(abs(out))
axs[1].plot(fv)
sp=fftshift(out)*2/len(xs)  ### scalling to the right amplitudes
freq=fftshift(fv)
plt.figure()
plt.plot(freq,abs(sp))
```
