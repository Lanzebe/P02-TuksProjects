## 1.5. Give expressions for the displacement field (x-displacement field u(x,y) and y-displacement field v(x,y)) for the indicated deformation: [4]
## You may assume the form: $u(x,y) = ax + bx + cxy + d$ and $v(x,y) = ex + fx + gxy + h$
### Example of solving a linear system of equations

```python
print('Answer here')
import numpy as np
initail_points = np.array([[0,0],[5,0],[0,10],[5,10]])
final_points = np.array([[0,0],[10,0],[1,10],[4,10]])
delta_points = final_points - initail_points


def dis_field_u(p):
    return np.array([p[0],p[1],p[0]*p[1],1])

def dis_field_v(p):
    return np.array([p[0],p[1],p[0]*p[1],1])

x_answer = np.array([delta_points[0][0],delta_points[1][0],delta_points[2][0],delta_points[3][0]])
y_answer = np.array([delta_points[0][1],delta_points[1][1],delta_points[2][1],delta_points[3][1]])

matrix_x = np.array([dis_field_u(initail_points[0]),dis_field_u(initail_points[1]),dis_field_u(initail_points[2]),dis_field_u(initail_points[3])])
matrix_y = np.array([dis_field_v(initail_points[0]),dis_field_v(initail_points[1]),dis_field_v(initail_points[2]),dis_field_v(initail_points[3])])

#print(matrix_x)
#print(x_answer)
#print(matrix_y)
#print(y_answer)

ans_u = np.linalg.solve(matrix_x,np.transpose(x_answer))
ans_v = np.linalg.solve(matrix_y,np.transpose(y_answer))

print('a:',ans_u[0])
print('b:',ans_u[1])
print('c:',ans_u[2])
print('d:',ans_u[3])

print('e:',ans_v[0])
print('f:',ans_v[1])
print('g:',ans_v[2])
print('h:',ans_v[3])
```

## 2.3 Compute the strain tensor from the given displacement field and give the engineering shear strain $\gamma_{xy}$ i.e. $\gamma_{xy} = \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x}$  [2]

## 3.1. Compute the strain matrix (tensor) given the displacement field below: [4] 
### Example of sympy.derive_by_array
### Example of transpose operator
### Example of adding two matrices

```python
import sympy
rho, g, nu, E, W, B, H, x, y, z = sympy.symbols('rho, g, nu, E, W, B, H, x, y, z')

u = -nu*rho*g*x*z/E
v = -nu*rho*g*y*z/E
w = rho*g*z**2/(2*E) + nu*rho*g*(x**2+y**2)/(2*E) - rho*g*H**2/(2*E)
#----------------------------------------------------

U = [u,v,w]
grad_of_U = sp.derive_by_array(U,[x,y,z]).simplify()
strainM  = 0.5* (grad_of_U + grad_of_U.transpose())
display(strainM)
```

## 3.2. Compute the stress matrix (tensor) the strain matrix computed before: [4]
## The 6x6 material matrix D is already given in symbolic format
### Example of converting a strain matrix into a strain vector - Voigt notation
### Example of multiplying two matrices

```python
# Note here D is defined as a sympy Matrix - this allows for multiplication by E/((1 + nu)*(1 - 2*nu)) as well as
# Linear Algebra multiplication when considering other matrices
# Note Voigt Notation for Material Matrix - Requires 6x1 Strain Vector
D = E/((1 + nu)*(1 - 2*nu))*sympy.Matrix([[1-nu,nu,nu,0,0,0],[nu,1-nu,nu,0,0,0],[nu,nu,1-nu,0,0,0],[0,0,0,0.5-nu,0,0],[0,0,0,0,0.5-nu,0],[0,0,0,0,0,0.5-nu]])

strain = sympy.Matrix([strainM[0][0],strainM[1][1],strainM[2][2],strainM[0][1],strainM[1][2],strainM[2][0]])

stress = sp.simplify(D*strain)
display(stress)
```
