```python
import matplotlib.pyplot as plt
import numpy as np

L = 1.5 #(m)
E = 210E9 #(Pa)
I = 1/12*(25/1000)*(50/1000)**3 #(m**4)
q = 1000 #(N/m)

x = np.linspace(0,L)

#v = lambda x: q*x**2/(24*E*I)*(-x**2 + 4*L*x - 6*L**2) # One way to define a function
def v(x):  # Antoher way to define a function
    return q*x**2/(24*E*I)*(-x**2 + 4*L*x - 6*L**2)

plt.plot(x,v(x),'r',label='deformed')
plt.xlabel('x [m]',fontsize=24)
plt.ylabel('v(x) (m)',fontsize=24)
plt.legend()
plt.show()
```
