# Question 1 [17]
# Consider the prismatic bar hanging under its own weight i.e. under gravity with a body force of $\rho g$ N/m$^3$. The bar has a height of $H$, width $2W$ and breadth of $2B$ as indicated, and is made of a material with Young's modulus E and Poisson's ratio $\nu$. The bar is fully clamped at the top. Consider the coordinate system as indicated with the following displacement field, which depends on (x,y,z):
# Displacement in the x-direction: $u_x(x,y,z) = -\frac{\nu\rho g xz }{E}$
# Displacement in the y-direction: $u_y(x,y,z) = -\frac{\nu\rho g yz }{E}$
# Displacement in the z-direction: $u_z(x,y,z) = \frac{\rho g z^2}{2E} + \frac{\nu \rho g (x^2 + y^2)}{2E} - \frac{\rho g H^2}{2E}$

```python
from IPython.display import Image
Image('figures/bar_own_weight.png',embed=True)
```

# 1. Compute the strain matrix (tensor) given the displacement field below: [2] 

```python
import sympy
rho, g, nu, E, W, B, H, x, y, z = sympy.symbols('rho, g, nu, E, W, B, H, x, y, z')

u = -nu*rho*g*x*z/E
v = -nu*rho*g*y*z/E
w = rho*g*z**2/(2*E) + nu*rho*g*(x**2+y**2)/(2*E) - rho*g*H**2/(2*E)
```

# 2. Compute the stress from the strain matrix computed before: [4]
## The 6x6 material matrix D is already given in symbolic format

```python
D = E/((1 + nu)*(1 - 2*nu))*sympy.Matrix([[1-nu,nu,nu,0,0,0],[nu,1-nu,nu,0,0,0],[nu,nu,1-nu,0,0,0],[0,0,0,0.5-nu,0,0],[0,0,0,0,0.5-nu,0],[0,0,0,0,0,0.5-nu]])
```

# Question 2 [18]
# Consider the plane stress problem of a unit thickness $-0.5 \leq  z \leq 0.5$ structure with the Cartesian coordinate system given in terms of $x, y$ and $z$, depicted below.
# Recall plane stress problems imply there are zero surface tractions on the front and back (out of the paper along $z$-direction) surfaces.

```python
from IPython.display import Image
Image('figures/uniform_load.png',embed=True)
```

```python
import sympy as sp
import numpy as np

x,y,z,E,nu,I,c,w,l = sp.symbols('x,y,z,E,nu,I,c,w,l')

# Note "w" here is the distributed load and not the displacement in the z-direction

u = w/(2*E*I)*((l**2*x-x**3/3)*y 
                + x*(2*y**3/3 - 2*c**2*y/5) 
                + nu*x*(y**3/3 - c**2*y + 2*c**3/3)); print('u:')#;sp.pretty_print(u)
display(u)

v = -w/(2*E*I)*(y**4/12 - c**2*y**2/2 + 2*c**3*y/3 
                 + nu*((l**2 - x**2)*y**2/2 + y**4/6 - c**2*y**2/5) 
                 -x**4/12 
                 + x**2*(l**2/2 + c**2*(4/5 + nu/2))) + 5*w*l**4/(24*E*I)*(1 + 12*c**2/(5*l**2)*(4/5+nu/2)); print('v:')#;sp.pretty_print(v)
display(v)
```

# 1. Compute the strain matrix (tensor) given the displacement field below: [2] 

```python
# case of PLANE STRESS; Sigma_zz = tau_xz = Tau_yz = 0

displacement_gradient =  sp.derive_by_array([u,v],[x,y]) #;display(displacement_gradient)
strain_tensor_partial = 0.5*( displacement_gradient + displacement_gradient.transpose() ) ;strain_tensor_partial=strain_tensor_partial.simplify() ; display('strain_tensor_partial=',strain_tensor_partial)
# Voight notation, using plane stress knowledge: Gamma_yz = Gamma_xz = 0, Epsilon_zz = -(nu*Epsilon_xx+nu*Epsilon_yy)/(1-nu)
Epsilon_xx,Epsilon_yy,Gamma_xy = strain_tensor_partial[0,0],strain_tensor_partial[1,1],2*strain_tensor_partial[0,1]
Epsilon_zz = -(nu*Epsilon_xx+nu*Epsilon_yy)/(1-nu)
Gamma_yz,Gamma_xz = 0,0 # 3d: strain_vector = sp.Matrix([Epsilon_xx,Epsilon_yy,Epsilon_zz,Gamma_yz,Gamma_xz,Gamma_xy])

strain_vector = sp.Matrix([Epsilon_xx,Epsilon_yy,Gamma_xy]) ; display('strain vector =',strain_vector) 
strain_tensor = sp.Matrix([[Epsilon_xx,Gamma_xy],[Gamma_xy,Epsilon_yy]]) ; display('strain tensor =',strain_tensor) 
```

# 2. Compute the stress-tenosr (tensor) from the strain matrix computed before: [4]
## Given are two  3x3 material matrice, D1 and D2, already given in symbolic format. One of them is for plane stress, while the other is for plane strain.

```python
D1 = E/(1-nu**2)*sp.Matrix([[1,nu,0],[nu,1,0],[0,0,(1-nu)/2]]);print('D1: Plane stress'); display(D1)#;sp.pretty_print(D1)
D2 = E/((1+nu)*(1-2*nu))*sp.Matrix([[1-nu,nu,0],[nu,1-nu,0],[0,0,(1-2*nu)/2]]);print('D2: Plane strain'); display(D2)#;sp.pretty_print(D2)

#Plane stress -> D1
stress_vector = D1*strain_vector ; stress_vector = sp.simplify(stress_vector)
print('stress vector');display(stress_vector)

stress_tensor = sp.Matrix([[stress_vector[0],stress_vector[2]],[stress_vector[2],stress_vector[1]]]) 
print('stress-tenosr (tensor)');display(stress_tensor)
```

# 3. Check whether the computed stress field is in equilibrium: [6]

```python
# for stress equilibrium: sum forces (with stress)=0, Plane stress-> no forces in z
# sum forces in x =0: No body forces in x-direction
sigma_dx = sp.derive_by_array(stress_vector,x) #; display(sigma_dx)
sum_x_F = sigma_dx[0] + sigma_dx[2] ; print('for stress equilibrium in x-direction, Sum of forces should be zero:',sum_x_F[0])
# sum forces in y =0: Body forces in y-direction
sigma_dx = sp.derive_by_array(stress_vector,y) #; display(sigma_dx)
body_force_y = 2*w*l - w*(2*l) ; print('\nbody forces in y-direction:',body_force_y) 
sum_y_F = sigma_dx[1] + sigma_dx[2]  ; print('for stress equilibrium in y-direction, Sum of forces should be zero:',sum_y_F[0])
```

```python
'''As seen above the computed stress fields are not in equilibrium as the x and y equilibrium sum of derivatives of stress are not equal to zero'''
```

# 4. Motivate whether the prescribed surface tractions for the front and back (out of the page along z-direction) are satisfied: [2]

```python
print('For plane stress state to be satisfied, it is prescribed that the surface tractions should be zero'
      ' on the front and back surfaces. '
      '\nChecking: the stress in the z-direction should be zero. Computed stress components exist for x and y only, i.e. in plane only with zero z component.'
     )
display("stress vector",stress_tensor)
print('Thus surface traction for front and back satisfied')
```

# 5. Check whether the prescribed surface tractions for top and bottom are satisfied: [4]

```python
# at top y = -c
print('stress_top:') ; stress_top = stress_tensor.subs(y,-c)
display(stress_top)
traction_top = stress_top* sp.Matrix( [ [0],[-1] ] ) ; print('traction vector at top surface:');display(traction_top)
print('traction vector at top surface has only a y component as expected, which is due to body forces')

# at bottom y = c
print('\n\nstress_bottom:') ; stress_bottom = stress_tensor.subs(y,c)
display(stress_bottom)
traction_bottom = stress_bottom* sp.Matrix( [ [0],[1] ] ) ; print('traction vector at bottom surface:');display(traction_bottom)
print('traction vector at bottom surface has only zero components since no body forces at the bottom of the structure')
```

# 5. Check whether the prescribed loading for left and right surfaces are satisfied: [6]
## The loading of the left and right surfaces are given as an integral of the surface tractions on the left and right surfaces, hence the units of $wl$ is N/m (integration along only one direction) or N (integration both y and z directions) and not N/m2. This implies we need to integrate the computed surface tractions along a direction. Integration needs to be performed along the y-direction.
## The integral for expression E for variable y between -c and c can be performed using sp.integrate(E,(y,-c,c)) - type help(sp.integrate) for additional information

```python
print('stress_left:') ; stress_left = stress_tensor.subs(x,-l)
display(stress_left)
traction_left = stress_left* sp.Matrix( [ [-1],[0] ] ) ; print('traction vector at left surface:');display(traction_left)
loading_left_surface = sp.integrate(traction_left,(y,-c,c)) ; print('loading at left surface:');display(loading_left_surface)
'''At left surface only loading in the y as expected, this is due to the body force wl, direction is upwards (-ve sign) as expected'''

print('\n\nstress_right:') ; stress_right = stress_tensor.subs(x,l)
display(stress_right)
traction_right = stress_right* sp.Matrix( [ [1],[0] ] ) ; print('traction vector at right surface:');display(traction_right)
loading_right_surface = sp.integrate(traction_right,(y,-c,c)) ; print('loading at right surface:');display(loading_right_surface)
'''At right surface only loading in the y as expected, this is due to the body force wl, direction is upwards (-ve sign) as expected'''
```

# Discuss whether the prescribed displacements are satisfied [2]

```python
'''Prescribed discplacement varies in x and y with no displacement prescribed in the z. 
For plane stress some displacement in the z is expected due to poisson contraction '''
'''As seen in the figure below the '''
```

```python
import numpy as np
import matplotlib.pyplot as plt

u = w/(2*E*I)*((l**2*x-x**3/3)*y 
                + x*(2*y**3/3 - 2*c**2*y/5) 
                + nu*x*(y**3/3 - c**2*y + 2*c**3/3))

v = -w/(2*E*I)*(y**4/12 - c**2*y**2/2 + 2*c**3*y/3 
                 + nu*((l**2 - x**2)*y**2/2 + y**4/6 - c**2*y**2/5) 
                 -x**4/12 
                 + x**2*(l**2/2 + c**2*(4/5 + nu/2))) + 5*w*l**4/(24*E*I)*(1 + 12*c**2/(5*l**2)*(4/5+nu/2))

constants = {'E': 210E9, #(Pa)
             'nu': 0.3,
             'c': 0.5, # m
             'w': 1, # m
             'l': 1.5, # m
             }
I = 1/12*1*(2*c)**3

x1 = np.linspace(-constants['l'],constants['l'],15)
y1 = np.linspace(-constants['c'],constants['c'],15) 

u = u.subs(constants) ; u=u.simplify() ; print('u(x,y) =') ; display(u)
v = v.subs(constants) ; v=v.simplify() ; print('v(x,y) =') ; display(v)

X1,Y1 = np.meshgrid(x1,y1)
lam_u = sp.lambdify([x, y],u,modules = 'numpy') ; u_disp = lam_u(X1, Y1)
lam_v = sp.lambdify([x, y],v,modules = 'numpy') ; v_disp = lam_v(X1, Y1)

fig = plt.figure()
ax = fig.add_subplot()
ax.invert_yaxis()
# ax.invert_xaxis()
plt.quiver(X1,Y1,u_disp,-v_disp)
plt.xlabel('Displacement, x-coordinate [m]',fontsize = 15)
plt.ylabel('Displacement, y-coordinate [m]',fontsize = 15)

plt.show()
```
