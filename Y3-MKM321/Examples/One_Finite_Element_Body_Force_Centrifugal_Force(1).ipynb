# Problem 1: Adapted from Example 3.7 Prescribed Textbook
## Instead of solving this problem with two three-noded elements like in the prescribed textbook - we will solve it with only one three-node element.
## Prescribed textbook example applies a uniform average body force over each element instead of a varying body force over each element.  Here we will compute the CONSISTENT nodal loads for a linerly varying body force over the element!
## Consider Figure E3.7: angular velocity of $\omega = 30$ rad/s. 
## Centrifugal force points radially outward from the axis of rotation (that is why the rope with ball is taut when swinging it around) - along the direction of the bar i.e. along x-direction!
## In the example they work with $\rho$ being the weight density i.e. $\rho = \frac{mg}{V}$ with untis N/m^3
## The magnitude of the centrifugal force is given by $F_x(x) = m \omega^2 x = \frac{\rho}{g} V(x)\omega^2 x$.
## The magnitude of the centrifugal body force is therefore $f_x(x) = \frac{\rho}{g} \omega^2 x$.
## Body force is linear in $x$
## The exact stress solution is given as $\sigma_{xx_{exact}} = \frac{\rho\omega^2}{2g}(L^2 - x^2)$
## In addition, the work done by the body force is equal to the strain energy: $\frac{1}{2}\int_V \sigma_{xx}\epsilon_{xx}dV$

```python
import numpy as np
import sympy as sp
xi = sp.symbols('xi')

# Define the shape functions
N1 = -1/2*xi*(1-xi)
N2 = 1/2*xi*(1+xi)
N3 = (1-xi)*(1+xi)

# Isoparametric formulation - relate x and xi
x1 = 0;x2 = 42;x3 = 21
x = N1*x1 + N2*x2 + N3*x3

# Define the area in terms of x, which in turn is related to xi - here A(xi) since x(xi).
A = 0.6+0*xi #in^2 the 0*xi term is to convert the type for A
print('A: {}'.format(A.simplify()))

E = 10E7 # psi
g_feet = 32.2 # feet/s^2
g = g_feet*12 # in/s^2 here in inches
rho = 0.2836 # weight density
omega = 30
L = x2-x1

# Compute non-zero surface traction
fx = rho*x*omega**2/g

# Setup a list of the shape functions N1, N2 and N3
N = [N1,N2,N3]

# Stiffness matrix
dxdxi = sp.derive_by_array(x,xi); print('dxdxi {}'.format(dxdxi))
dxidx = 1/dxdxi

B = dxidx*sp.derive_by_array(N,xi); print('B {}'.format(B))
print('B here is a vector when considering the shape {} '.format(B.shape))

print('If we want to levarage the multiplication symbol * to do linear Algebra multiplication we can cast B into a column vector 3x1 and then take the tranpose to get a row vector')
print('sympy.Matrix support for shape=(1,3) has unfortunately not been incorporated!')
print('Hence to convert a vector into a (1,3) matrix we need to:' )
print('Convert vector into sympy.Matrix has shape {} i.e. 3 rows and 1 column'.format(sp.Matrix(B).shape))
print('Transpose recast it as a 1x3 matrix with shape {}'.format(sp.Matrix(B).transpose().shape))
print('Hence we convert vector into a 3x1 matrix, when we convert into a 1x3 matrix')
Bmat = sp.Matrix(B).transpose() # sp.Matrix converts a 3 vector (3) into a column vector (3x1) by default!s
BTEB = Bmat.transpose()*E*Bmat; print('BT*E*B whichs is a 3x1 vector times a 1x3 vector {}'.format(BTEB))

integrandStiffnessMatrix = BTEB*A*dxdxi
# 2-point Gauss Quadrature Integration
StiffnessMatrix = 1.0*integrandStiffnessMatrix.subs({xi:-1/3**0.5}) + 1.0*integrandStiffnessMatrix.subs({xi:1/3**0.5})

# Body force vector that results in force is given by
Fintegrand = fx*A*sp.Matrix(N)*dxdxi
# 2-point Gauss Quadrature Integration - highest order term in integrand is 3 (shape function quadratic and body force is linear)
F = 1.0*Fintegrand.subs({xi:-1/3**0.5}) + 1.0*Fintegrand.subs({xi:1/3**0.5}); print('Body Force Vector {}'.format(F))

#The first row is multiplied by W1=0, hence only row 2&3 remains, we move the first column to the traction side 
StiffnessMatrixReduced = StiffnessMatrix[1:,1:]
#Matrix has a built-in method solve, we need to only specify the Right-Hand-Side of the Linear System to Solve
solution = StiffnessMatrixReduced.solve(F[1:,0]-StiffnessMatrix[1:,0]*0); print('Solution for q2 and q3 is {}'.format(solution))

# Here we override 
q1 = 0 # prescribed displacement
q2 = solution[0,0] #solution from 2x2 Matrix system which is a 2x1 vector
q3 = solution[1,0] #solution from 2x2 Matrix system which is a 2x1 vector
q = sp.Matrix([q1,q2,q3])

# Plotting the result - here we levarage built-in dot operator for the sympy.Matrix object
u = sp.Matrix(N).dot(q) # Matrix with shape functions dot with solution vector
sp.plot(u,(xi,-1,1),title="Displacement")

strain = Bmat.dot(q) # B matrix dot with solution vector
sp.plot(strain,(xi,-1,1),title="Strain")

stress = E*strain
stress_exact = rho*omega**2/(2*g)*(L**2 - x**2)
strain_exact = stress_exact/E
sp.plot(stress,stress_exact,(xi, -1, 1),title="Stress vs Stress Exact");

internalforce = stress*A
internalforce_exact = stress_exact*A
sp.plot(internalforce,internalforce_exact,(xi,-1,1),title="Internal Force vs Internal Force Exact")

strainfirstmodel = strain # to be used for the next model when we compute the strain energy

print('---------STRESS-----------')
print('The largest FEM predicted stress is at the rotation point is {} psi which is {} % larger than the exact stress {} psi'.format(stress.subs({'xi':-1}),
                                                                                                                    (stress.subs({'xi':-1})-stress_exact.subs({xi:-1}))/stress_exact.subs({'xi':-1})*100,
                                                                                                                    stress_exact.subs({'xi':-1})))
print('The stress at the tip is expected to be 0 but it is {} psi'.format(stress.subs({'xi':1})))

print('---------WORK EXACT ESTIMATE COMPARISON (STRESS EXACT, STRAIN APPROXIMATE) -----------')
print('Lets estimate the strain energy done by the centrifugal force in the FEM:') 
Work = sp.integrate(0.5*stress*strain*A*dxdxi,(xi,-1,1)); print('FEM Strain Energy: Integral FEM stress times FEM strai: {}'.format(Work))
WorkExactEstimate = sp.integrate(0.5*stress_exact*strainfirstmodel*A*dxdxi,(xi,-1,1)); print('Exact Estimate Strain Energy: Integral Exact Estimate Integral exact stress times FEM strain: {}'.format(WorkExactEstimate))
print('Difference in Work between the FEM and Estimated Exact Solution is {} %'.format(abs(Work-WorkExactEstimate)/WorkExactEstimate*100))
print('Consistent Nodal Load Body Force ensures that the Strain Energy between the Estimated Exact and FEM solutions is the same!')
print('Work done by the centrifugal force for the FEM solution is same as the estimated exact solution!')
print('Meaning: Consistent nodal load tractions give energy equivalent to exact stress solution!')

print('---------WORK EXACT COMPARISON (STRESS EXACT, STRAIN EXACT)-----------')
print('Lets estimate the strain energy done by the centrifugal force in the FEM:') 
Work = sp.integrate(0.5*stress*strain*A*dxdxi,(xi,-1,1)); print('FEM Strain Energy: Integral FEM stress times FEM strai: {}'.format(Work))
WorkExact = sp.integrate(0.5*stress_exact*strain_exact*A*dxdxi,(xi,-1,1)); print('Exact Estimate Strain Energy: Integral Exact Estimate Integral exact stress times FEM strain: {}'.format(WorkExact))
print('Difference in Work between the FEM and Exact Solution is {} %'.format(abs(Work-WorkExact)/WorkExact*100))
print('Consistent Nodal Load Body Force ensures that the Strain Energy between the Exact and FEM solutions is similar!')
print('Work done by the centrifugal force for the FEM solution is close to the exact solution!')
```

# Problem 1: Example 3.7 Prescribed Textbook (single element instead of two)
## Prescribed textbook example applies a uniform average body force over each element instead of a varying body force over each element.  Now we will apply only the ELEMENT AVERAGE BODY FORCE uniformly over the element i.e. it is not modelled as linearly varying but rather as a constant!
## Consider Figure E3.7: angular velocity of $\omega = 30$ rad/s. 
## Centrifugal force points radially outward from the axis of rotation (that is why the rope with ball is taut when swinging it around) - along the direction of the bar i.e. along x-direction!

## The magnitude of the centrifugal body force is therefore $f_x(x) = \frac{\rho}{g} \omega^2 x$.
## Body force is linear in $x$ but modelled as constant over $x$ using the AVERAGE BODY FORCE

```python
# Similar to prescribed textbook example - compute body force in middle of element 
# (here bar since we only have one element) and assume it is uniform over the entire bar.
fx = rho*(L/2)*omega**2/g

# Setup a list of the shape functions N1, N2 and N3
N = [N1,N2,N3]

# Stiffness matrix
dxdxi = sp.derive_by_array(x,xi); print('dxdxi {}'.format(dxdxi))
dxidx = 1/dxdxi

B = dxidx*sp.derive_by_array(N,xi); print('B {}'.format(B))
print('B here is a vector when considering the shape {} '.format(B.shape))

print('If we want to levarage the multiplication symbol * to do linear Algebra multiplication we can cast B into a column vector 3x1 and then take the tranpose to get a row vector')
print('sympy.Matrix support for shape=(1,3) has unfortunately not been incorporated!')
print('Hence to convert a vector into a (1,3) matrix we need to:' )
print('Convert vector into sympy.Matrix has shape {} i.e. 3 rows and 1 column'.format(sp.Matrix(B).shape))
print('Transpose recast it as a 1x3 matrix with shape {}'.format(sp.Matrix(B).transpose().shape))
print('Hence we convert vector into a 3x1 matrix, when we convert into a 1x3 matrix')
Bmat = sp.Matrix(B).transpose() # sp.Matrix converts a 3 vector (3) into a column vector (3x1) by default!s
BTEB = Bmat.transpose()*E*Bmat; print('BT*E*B whichs is a 3x1 vector times a 1x3 vector {}'.format(BTEB))

integrandStiffnessMatrix = BTEB*A*dxdxi
# 2-point Gauss Quadrature Integration
StiffnessMatrix = 1.0*integrandStiffnessMatrix.subs({xi:-1/3**0.5}) + 1.0*integrandStiffnessMatrix.subs({xi:1/3**0.5})

# Body force vector that results in force is given by
Fintegrand = fx*A*sp.Matrix(N)*dxdxi
# 2-point Gauss Quadrature Integration - highest order term in integrand is 2 (shape function quadratic and body force is modelled as constant)
F = 1.0*Fintegrand.subs({xi:-1/3**0.5}) + 1.0*Fintegrand.subs({xi:1/3**0.5}); print('Body Force Vector {}'.format(F))

#The first row is multiplied by W1=0, hence only row 2&3 remains, we move the first column to the traction side 
StiffnessMatrixReduced = StiffnessMatrix[1:,1:]
#Matrix has a built-in method solve, we need to only specify the Right-Hand-Side of the Linear System to Solve
solution = StiffnessMatrixReduced.solve(F[1:,0]-StiffnessMatrix[1:,0]*0); print('Solution for q2 and q3 is {}'.format(solution))

# Here we override 
q1 = 0 # prescribed displacement
q2 = solution[0,0] #solution from 2x2 Matrix system which is a 2x1 vector
q3 = solution[1,0] #solution from 2x2 Matrix system which is a 2x1 vector
q = sp.Matrix([q1,q2,q3])

# Plotting the result - here we levarage built-in dot operator for the sympy.Matrix object
u = sp.Matrix(N).dot(q) # Matrix with shape functions dot with solution vector
sp.plot(u,(xi,-1,1),title="Displacement")

strain = Bmat.dot(q) # B matrix dot with solution vector
sp.plot(strain,(xi,-1,1),title="Strain")

stress = E*strain
stress_exact = rho*omega**2/(2*g)*(L**2 - x**2)
sp.plot(stress,stress_exact,(xi, -1, 1),title="Stress vs Stress Exact")

internalforce = stress*A
sp.plot(internalforce,internalforce_exact,(xi,-1,1),title="Internal Force vs Internal Force Exact")


print('---------STRESS-----------')
print('The largest FEM predicted stress is at the rotation point is {} psi which is {} % difference to the exact exact stress {} psi - maximum stress is exactly predicted'.format(stress.subs({'xi':-1}),
                                                                                                                    (stress.subs({'xi':-1})-stress_exact.subs({xi:-1}))/stress_exact.subs({'xi':-1})*100,
                                                                                                                    stress_exact.subs({'xi':-1})))
print('The stress at the tip is is indeed 0, numerically given by {} psi'.format(stress.subs({'xi':1})))

print('---------WORK EXACT ESTIMATE COMPARISON (STRESS EXACT, STRAIN APPROXIMATE) -----------')
print('Lets estimate the strain energy done by the centrifugal force in the FEM:') 
Work = sp.integrate(0.5*stress*strain*A*dxdxi,(xi,-1,1)); print('FEM Strain Energy: Integral FEM stress times FEM strai: {}'.format(Work))
WorkExactEstimate = sp.integrate(0.5*stress_exact*strainfirstmodel*A*dxdxi,(xi,-1,1)); print('Exact Estimate Strain Energy: Integral Exact Estimate Integral exact stress times FEM strain: {}'.format(WorkExactEstimate))
print('Difference in Work between the FEM and Estimated Exact Solution is {} %'.format(abs(Work-WorkExactEstimate)/WorkExactEstimate*100))
print('Approximating body force as piece-wise constant instead of linearly varying results in an under-prediction of the Strain Energy of the FEM solution!')

print('---------WORK EXACT COMPARISON (STRESS EXACT, STRAIN EXACT)-----------')
print('Lets estimate the strain energy done by the centrifugal force in the FEM:') 
Work = sp.integrate(0.5*stress*strain*A*dxdxi,(xi,-1,1)); print('FEM Strain Energy: Integral FEM stress times FEM strai: {}'.format(Work))
WorkExact = sp.integrate(0.5*stress_exact*strain_exact*A*dxdxi,(xi,-1,1)); print('Exact Estimate Strain Energy: Integral Exact Estimate Integral exact stress times FEM strain: {}'.format(WorkExact))
print('Difference in Work between the FEM and Exact Solution is {} %'.format(abs(Work-WorkExact)/WorkExact*100))
print('Approximating body force as piece-wise constant instead of linearly varying results in an under-prediction of the Strain Energy of the FEM solution!')
```
