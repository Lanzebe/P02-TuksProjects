```python
from ansys.mapdl.core import launch_mapdl
import re
import numpy as np
from random import uniform

path = 'c:\\Users\\ebers\\Simulations\\MRN412\\Testing Scripting'
mapdl = launch_mapdl(run_location=path, jobname='TestingScripting',override=True)


```

```python
data = [[0.01,0.1525,0.107,70E9,2700,0.33],
        [0.006,0.2935,0.1305,70E9,2700,0.33],
        [0.006,0.3265,0.09,70E9,2700,0.33],
        [0.008,0.189,0.101,70E9,2700,0.33],
        [0.010,0.1844,0.152,70E9,2700,0.33],
        [0.01,0.164,0.122,70E9,2700,0.33],
        [0.01,0.1595,0.0917,70E9,2700,0.33]]

reset = True

if reset:
    RawDataToBeSaved = np.empty((0,14),float)
else:
    RawDataToBeSaved = np.load('RawDataToBeSaved.npy', allow_pickle=True)


for i in data:

    Core_data = i
    
    ## General Setup
    mapdl.clear()
    mapdl.prep7()
    mapdl.units("SI")  # SI - International system (m, kg, s, K).

    ##Geometry
    mapdl.block(0,i[1],0,i[2],0,i[0])
    #mapdl.vplot('all')


    ## Material properties
    mapdl.lesize('ALL',i[0], layer1=1) ### check seeds.

    mapdl.mp('ex',1,i[3])
    mapdl.mp('nuxy',1,i[5])
    mapdl.mp('dens',1,i[4])

    mapdl.et(1, 186)
    mapdl.esize(i[0])
    mapdl.vsweep("all")


    mapdl.mshape(1,'3D')
    mapdl.mshkey(0)
    mapdl.vmesh('all')
    #mapdl.eplot()

    mapdl.run("/SOLU")
    mapdl.antype("MODAL")
    output = mapdl.modal_analysis(nmode=8, freqb=1)

    Frequencies = []

    mapdl.post1()
    for n in range(8):
        resp = mapdl.set(lstep='1',sbstep=str(n+1))
        Frequencies.append(float(resp[153:159]))

    #print(Core_data)
    #RawDataToBeSaved = np.append(RawDataToBeSaved,np.array([Core_data]), axis=0)
    
    print(np.round(Frequencies))

    #if ((i+1)%50)==0:
        #np.save('RawDataToBeSaved',RawDataToBeSaved)

    
```
