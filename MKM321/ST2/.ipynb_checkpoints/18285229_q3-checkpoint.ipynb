# 3.1. Area relationship first element [2]
### Answer here

```python
import sympy as sp

xi = sp.symbols('xi')

A1 = 0.004
A2 = 0.003
A3 = 0.002


L0 = -1/2*xi*(1-xi)
L1 = 1/2*xi*(1+xi)
L2 = (1-xi)*(1+xi)

A = xpos = L0*A1 + L1*A3 + L2*A2

print('Area of first element as a function of xi:')
display(A)
print('Testing at xi=-1:')
print(A.subs({xi:-1}),'[mm2]')
print('Testing at xi=0:')
print(A.subs({xi:0}),'[mm2]')
print('Testing at xi=1:')
print(A.subs({xi:1}),'[mm2]')
print('All makes sense and is verified.')
```

# 3.2. Area relationship second element [2]
### Answer here

```python
import sympy as sp

xi = sp.symbols('xi')

A3 = 0.002
A4 = 0.003
A5 = 0.004


L0 = -1/2*xi*(1-xi)
L1 = 1/2*xi*(1+xi)
L2 = (1-xi)*(1+xi)

A = xpos = L0*A3 + L1*A5 + L2*A4

print('Area of second element as a function of xi:')
display(A)
print('Testing at xi=-1:')
print(A.subs({xi:-1}),'[mm2]')
print('Testing at xi=0:')
print(A.subs({xi:0}),'[mm2]')
print('Testing at xi=1:')
print(A.subs({xi:1}),'[mm2]')
print('All makes sense and is verified.')
```

# 3.3. $x$-value for given $xi$ value [2]:
### Answer here

```python
import sympy as sp

xi = sp.symbols('xi')

x1 = 0
x2 = 75
x3 = 150


L0 = -1/2*xi*(1-xi)
L1 = 1/2*xi*(1+xi)
L2 = (1-xi)*(1+xi)

X = xpos = L0*x1 + L1*x3 + L2*x2

print('X_position of first element as a function of xi:')
display(X)
print('Testing at xi=-1:')
print(X.subs({xi:-1}),'[mm]')
print('Testing at xi=0:')
print(X.subs({xi:0}),'[mm]')
print('Testing at xi=1:')
print(X.subs({xi:1}),'[mm]')
print('So the value of X at xi=0.25 will be:')
print(X.subs({xi:0.25}),'[mm]')
```
