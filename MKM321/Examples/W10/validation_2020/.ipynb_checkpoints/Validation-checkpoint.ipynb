## Isoclinics 
### Isoclinics are black fringes that join all points with the same principal stress directions are the same 
$$\tilde{\tau}_{xy} =(\sigma_{yy}-\sigma_{xx})\sin\theta\cos\theta + \tau_{xy}(cos^2\theta - sin^2\theta)$$
### Principal stress orientation $\theta_P$, occurs when $\tilde{\tau}_{xy} = 0,$, hence
$$0 =(\sigma_{yy}-\sigma_{xx})\sin\theta_P\cos\theta_P + \tau_{xy}(cos^2\theta_P - sin^2\theta_P)$$

$$\tan(2\theta_P) = \frac{2\tau_{xy}}{\sigma_{xx}-\sigma_{yy}}$$
## Isochromatics
### Isochromatics are lines of constant colour when a white light source are used. The colours are related to the level of loading, specifically contours of constant principal stress. Stress-optic law: $$\sigma_1 - \sigma_2 = f_\sigma \frac{N}{t},$$ where $\sigma_1$ and $\sigma_2$ are the principal stresses at a point, $f_\sigma$ material fringe constant (material property), $N$ the fringe order and $t$ the thickness of the structure.
