# Volumetric Locking
## Four noded quadrilaterial with full integration
## Only for plane strain, when the strain energy associated with element deformation becomes artificially high when Poisson's ratio $\nu\rightarrow 0.5$
## Reminder $\nu\rightarrow 0.5$ the material approaches an incompressible material
## Why only plane strain - as plane stress has freedom to have strain of the plane, whereas in plane strain all strain occurs in the plane.

```python
from IPython.display import Image
print('Credit to Wikipedia for this Table')
Image("material_proterty_relationships.png")
```

```python
import numpy as np
E = 2.1E5
v = 0.49

# Material Matrices
matDpstress = E/(1-v**2)*np.array([[1,v,0],[v,1,0,],[0,0,(1-v)/2]])
matDpstrain = E/((1+v)*(1-2*v))*np.array([[1-v,v,0],[v,1-v,0,],[0,0,(1-2*v)/2]])
print('-----------------------')

matD = matDpstrain

import numpy as np
from elements import Quad4_Stiffness

X = np.array([[0.0, -1],
              [2.0, -1],
              [2.0,  1],
              [0.0,  1]])



Kelem = Quad4_Stiffness(X,matD,1.0)

Eigenvalue, Eigenvector = np.linalg.eig(Kelem)

print('Eight Eigenvalues {}'.format(Eigenvalue))
print('Each column in Eigenvector matrix is an eigenvector')

Mode_Index = 1
Deformation = Eigenvector[:,Mode_Index]
StrainEnergy = 0.5*np.dot(Deformation.transpose(),Kelem).dot(Deformation)

print('Eigenvalues {}'.format(Eigenvalue[Mode_Index]))
print('Check that Eigenvectors are Normalized {}'.format(np.linalg.norm(Deformation)))

import matplotlib.pyplot as plt
Factor = 0.5
NewX = (X[:,0] + Factor*np.take(Deformation,range(0,8,2))).transpose()
NewY = (X[:,1] + Factor*np.take(Deformation,range(1,8,2))).transpose()


plt.figure(100)
plt.plot(X[:,0],X[:,1],'k-o')
plt.plot(NewX, NewY,'r-o')
plt.title('Eigenvalue {} and Strain Energy {}'.format(Eigenvalue[Mode_Index],StrainEnergy))
plt.axis('equal')
plt.legend()
plt.show()
```
