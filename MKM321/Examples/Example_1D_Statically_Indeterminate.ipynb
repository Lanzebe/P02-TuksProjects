```python
import sympy as sp
```

```python
A1,E1,L1,A2,E2,L2,R1,R2,P,x = sp.symbols('A1,E1,L1,A2,E2,L2,R1,R2,P,x')
A,E,L = sp.symbols('A,E,L')
```

```python
Eq1 = P - R1 - R2
Eq2 = L1 + R1*L1/(A1*E1) + L2 - R2*L2/(A2*E2) - (L1+L2) 
result = sp.linsolve([Eq1,Eq2],[R1,R2]); print(result)

problem_details = {E1:10,E2:1,L1:1,L2:1,A1:1,A2:1,P:1}

result = result.subs(problem_details)
result_list = list(result)[0]

delta1 = R1*x/(A1*E1); delta1 = delta1.subs({R1:result_list[0]}); delta1 = delta1.subs(problem_details)
delta2 = -R2*(x-problem_details[L1])/(A2*E2); delta2 = delta2.subs({R2:result_list[1]}); delta2 = delta2.subs(problem_details)

sp.plotting.plot(delta1,(x,0,problem_details[L1]))
sp.plotting.plot(delta1.subs({x:problem_details[L1]})+delta2,(x,problem_details[L1],problem_details[L1]+problem_details[L2]))
```
