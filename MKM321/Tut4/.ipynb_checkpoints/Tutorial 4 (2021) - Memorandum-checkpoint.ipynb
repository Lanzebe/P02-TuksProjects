# Question 1 [17]
# Consider the prismatic bar hanging under its own weight i.e. under gravity with a body force of $\rho g$ N/m$^3$. The bar has a height of $H$, width $2W$ and breadth of $2B$ as indicated, and is made of a material with Young's modulus E and Poisson's ratio $\nu$. The bar is fully clamped at the top. Consider the coordinate system as indicated with the following displacement field, which depends on (x,y,z):
# Displacement in the x-direction: $u_x(x,y,z) = -\frac{\nu\rho g xz }{E}$
# Displacement in the y-direction: $u_y(x,y,z) = -\frac{\nu\rho g yz }{E}$
# Displacement in the z-direction: $u_z(x,y,z) = \frac{\rho g z^2}{2E} + \frac{\nu \rho g (x^2 + y^2)}{2E} - \frac{\rho g H^2}{2E}$

```python
from IPython.display import Image
Image('figures/bar_own_weight.png',embed=True)
```

# 1. Compute the strain matrix (tensor) given the displacement field below: [2] 

```python
import sympy
rho, g, nu, E, W, B, H, x, y, z = sympy.symbols('rho, g, nu, E, W, B, H, x, y, z')

u = -nu*rho*g*x*z/E
v = -nu*rho*g*y*z/E
w = rho*g*z**2/(2*E) + nu*rho*g*(x**2+y**2)/(2*E) - rho*g*H**2/(2*E)

# We want to compute the partial derivatives of displacement vector (ux,uy,uz) w.r.t. spatial variables (x,y,z)
displacement_vector = [u,v,w] # Note I add the components to a LIST - only the square brackets []
spatial_coordinates = [x, y, z] # Note I add the spatial components to a LIST - only the square brackets []

# Components inside lists when using derive_by_array
displacement_gradient = sympy.derive_by_array(displacement_vector, spatial_coordinates) 
print('Object type returned {} - sympy tensor array object, which supports linear algebra operations'.format(type(displacement_gradient)))

# Strain is defined as 0.5*(grad(u)^T + grad(u))
strain_matrix = 0.5*(displacement_gradient.transpose() + displacement_gradient)
sympy.pprint(strain_matrix)
print('Strain matrix - interpret the result: normal strain in x and y are Poisson contraction,\n due to gravity body force acting in the z direction')
```

# 2. Compute the stress from the strain matrix computed before: [4]
## The 6x6 material matrix D is already given in symbolic format

```python
# Note here D is defined as a sympy Matrix - this allows for multiplication by E/((1 + nu)*(1 - 2*nu)) as well as
# Linear Algebra multiplication when considering other matrices
# Note Voigt Notation for Material Matrix - Requires 6x1 Strain Vector
D = E/((1 + nu)*(1 - 2*nu))*sympy.Matrix([[1-nu,nu,nu,0,0,0],[nu,1-nu,nu,0,0,0],[nu,nu,1-nu,0,0,0],[0,0,0,0.5-nu,0,0],[0,0,0,0,0.5-nu,0],[0,0,0,0,0,0.5-nu]])

# Strain matrix needs to be converted into a strain vector
strain_vector = sympy.Matrix([strain_matrix[0,0],strain_matrix[1,1],strain_matrix[2,2],2*strain_matrix[1,2],2*strain_matrix[0,2],2*strain_matrix[0,1]])

stress_vector = D*strain_vector
sympy.pprint(stress_vector)
print('We can simplify the expressions')
stress_vector = sympy.simplify(stress_vector)
sympy.pprint(stress_vector)
print('Only one non-zero stress component that needs to balance out with body force')
```

# 3. Check whether the computed stress field is in equilibrium: [3]

```python
from sympy import derive_by_array

#stress vector [sxx, syy, szz, syz, sxz, sxy]
Eq1 = derive_by_array(stress_vector[0],x) + derive_by_array(stress_vector[5],y) + derive_by_array(stress_vector[4],z)
print('Equilibrium first equation: {}'.format(Eq1))

Eq2 = derive_by_array(stress_vector[5],x) + derive_by_array(stress_vector[1],y) + derive_by_array(stress_vector[3],z)
print('Equilibrium second equation: {}'.format(Eq2))

Eq3 = derive_by_array(stress_vector[4],x) + derive_by_array(stress_vector[3],y) + derive_by_array(stress_vector[2],z) - rho*g
print('Equilibrium third equation: {}'.format(Eq3))

print('Equilibrium is satisfied for each point x, y and z')
```

# 4. Check whether the prescribed surface tractions are satisfied: [5]

```python
print('On all surfaces there is a zero prescribed surface traction - we need to veryfy that this is indeed true')
print('To simplfiy calcs I will translate the stress vector into a stress matrix')
smatrix = sympy.Matrix([[stress_vector[0],stress_vector[5],stress_vector[4]],[stress_vector[5],stress_vector[1],stress_vector[3]],[stress_vector[4],stress_vector[3],stress_vector[2]]])
sympy.pprint(smatrix)
print('Recall the relationship stress_matrix(evaluated at surface to give plane) * normal = traction on surface')

print('Bottom surface - z-coordinate = 0 and outward pointing surface normal is [0,0,-1]')
nbottom = sympy.Matrix([0,0,-1])
print(smatrix.subs({z:0})*nbottom)

print('Front surface - x-coordinate = B and outward pointing surface normal is [1,0,0]')
nfront = sympy.Matrix([1,0,0])
print(smatrix.subs({x:B})*nfront)

print('Back surface - x-coordinate = -B and outward pointing surface normal is [-1,0,0]')
nback = sympy.Matrix([1,0,0])
print(smatrix.subs({x:-B})*nback)

print('Left surface - y-coordinate = -W and outward pointing surface normal is [0,-1,0]')
nleft = sympy.Matrix([0,-1,0])
print(smatrix.subs({y:-W})*nleft)

print('Right surface - y-coordinate = W and outward pointing surface normal is [0,1,0]')
nright = sympy.Matrix([0,1,0])
print(smatrix.subs({y:W})*nright)

print('All prescribed surface tractions are indeed [0,0,0] as specified')
```

# 5. Compute the reaction tractions resulting from the prescribed for the displacements and critically interpret: [3]

```python
print('Top surface - z-coordinate = H and outward pointing surface normal is [0,0,1]')
ntop = sympy.Matrix([0,0,1])
print(smatrix.subs({z:H})*ntop)

print('Reaction traction is upwards - required to handle gravity')
print('Reaction traction is upwards - required to handle gravity')
print('Reaction force top (integrate over x,y) = 1.0*H*g*rho*(2*W)*(2*B) = rho*(VOLUME)*g as expected')
```

# Question 2 [18]
# Consider the plane stress problem of a unit thickness $-0.5 \leq  z \leq 0.5$ structure with the Cartesian coordinate system given in terms of $x, y$ and $z$, depicted below.
# Recall plane stress problems imply there are zero surface tractions on the front and back (out of the paper along $z$-direction) surfaces.

```python
from IPython.display import Image
Image('figures/uniform_load.png',embed=True)
```

```python
import sympy as sp
import numpy as np

x,y,z,E,nu,I,c,w,l = sp.symbols('x,y,z,E,nu,I,c,w,l')

# Note w here is the distributed load and not the displacement in the z-direction

u = w/(2*E*I)*((l**2*x-x**3/3)*y 
                + x*(2*y**3/3 - 2*c**2*y/5) 
                + nu*x*(y**3/3 - c**2*y + 2*c**3/3)); print('u:');sp.pretty_print(u)

v = -w/(2*E*I)*(y**4/12 - c**2*y**2/2 + 2*c**3*y/3 
                 + nu*((l**2 - x**2)*y**2/2 + y**4/6 - c**2*y**2/5) 
                 -x**4/12 
                 + x**2*(l**2/2 + c**2*(4/5 + nu/2))) + 5*w*l**4/(24*E*I)*(1 + 12*c**2/(5*l**2)*(4/5+nu/2)); print('v:');sp.pretty_print(v)
```

# 1. Compute the strain matrix (tensor) given the displacement field below: [2] 

```python
displacement_gradient = sp.derive_by_array([u,v],[x,y])
epsilon = 0.5*(displacement_gradient.transpose() + displacement_gradient)
sp.pprint(epsilon)
```

# 2. Compute the stress-tensor(tensor) from the strain matrix computed before: [4]
## Given are two  3x3 material matrices, D1 and D2, already given in symbolic format. One of them is for plane stress, while the other is for plane strain.

```python
D1 = E/(1-nu**2)*sp.Matrix([[1,nu,0],[nu,1,0],[0,0,(1-nu)/2]]);print('D1:');sp.pretty_print(D1)
D2 = E/((1+nu)*(1-2*nu))*sp.Matrix([[1-nu,nu,0],[nu,1-nu,0],[0,0,(1-2*nu)/2]]);print('D2:');sp.pretty_print(D2)

epsvec = sp.Matrix([epsilon[0,0],epsilon[1,1],2*epsilon[0,1]])
D = D1  #........(1) Plane stress 2D
sigvec = D*epsvec #........(2)
print(sp.simplify(sigvec))

stress2D = sp.Matrix([[sigvec[0],sigvec[2]],[sigvec[2],sigvec[1]]]) # 2D Stress matrix
stress3D = sp.Matrix([[sigvec[0],sigvec[2],0],[sigvec[2],sigvec[1],0],[0,0,0]]) # 3D Stress matrix
```

# 3. Check whether the computed stress field is in equilibrium: [6]

```python
sig11 = sigvec[0]
sig22 = sigvec[1]
sig12 = sig21 = sigvec[2]
print('### Equilibrium ###')
dim1 = sp.diff(sig11,x)+sp.diff(sig21,y)
dim2 = sp.diff(sig12,x)+sp.diff(sig22,y)
print('Direction 1 : ',sp.simplify(dim1))
print('Direction 2 : ',sp.simplify(dim2))
print('Direction 3 : 0 due to plane stress problem',)

# 1pt = zero body force + zero acceleration
# 2pt = correct dervatives
# 2pt = correct summations

# -1 pt = check only one direction (either Direction_x or Direction_y)
```

# 3. Motivate whether the prescribed surface tractions for the front and back (out of the page along z-direction) are satisfied: [2]

```python
print('Plane stress - zero stress in z direction both front and back faces satisfied')
#    Plane stress. sigma33 = 0. Therefore,
print('t(x = -l:+l, y = -c:c, z = -0.5) = [0,0,0]')

### Back ###
#    Plane stress. sigma33 = 0. Therefore,
print('t(x = -l:+l, y = -c:c, z = +0.5) = [0,0,0]')
```

# 4. Check whether the prescribed surface tractions for top and bottom are satisfied: [4]

```python
sv = sigvec.applyfunc(sp.simplify)
sigmat = sp.Matrix([[sv[0],sv[2],0],[sv[2],sv[1],0],[0,0,0]])

### Top ###
print('Top Surface')
n = sp.Matrix([0,-1,0])
sp.pretty_print(sigmat.subs({'y':-c})*n)
print('I = 1/12*(2*c)**3')
sp.pretty_print(sigmat.subs({'y':-c,'I':1/12*(2*c)**3})*n)
print('It matches the applied traction at the top')

print('Bottom Surface')
n = sp.Matrix([0,1,0])
sp.pretty_print(sigmat.subs({'y':c})*n)
print('It matches the condition of zero applied traction at the bottom')

# 2pt = correct sigma (a function of x, y and z)
# 2pt = correct n (normal vectors)
# 2pt = all 6 sides
```

# 5. Check whether the prescribed loading for left and right surfaces are satisfied: [6]
## The loading of the left and right surfaces are given as an integral of the surface tractions on the left and right surfaces, hence the units of $wl$ is N/m (integration along only one direction) or N (integration both y and z directions). Integating along z (unit length) can be performed.

## The integral for expression E for variable y between -c and c can be performed using sp.integrate(E,(y,-c,c)) - type help(sp.integrate) for additional information

```python
print('Left Surface')
n = sp.Matrix([-1,0,0])
t = sigmat.subs({'x':-l})*n
t_surface_integral = sp.integrate(t,(y,-c,c)) # Unit thickness in z without variation.
t_surface_integral = sp.integrate(t_surface_integral,(z,-0.5,0.5)) # Unit thickness in z without variation.


sp.pretty_print(t_surface_integral)
print('I = 1/12*(2*c)**3')
sp.pretty_print(t_surface_integral.subs({'I':1/12*(2*c)**3}))
print('It matches the applied loading at the left surface -wl')

print('Right Surface')
n = sp.Matrix([1,0,0])
t = sigmat.subs({'x':l})*n
t_surface_integral = sp.integrate(t,(y,-c,c)) # Unit thickness in z without variation.
t_surface_integral = sp.integrate(t_surface_integral,(z,-0.5,0.5)) # Unit thickness in z without variation.

sp.pretty_print(t_surface_integral)
print('I = 1/12*(2*c)**3')
sp.pretty_print(t_surface_integral.subs({'I':1/12*(2*c)**3}))
print('It matches the applied loading at the right surface -wl')
```
