# Question 1 [17]
# Consider the prismatic bar hanging under its own weight i.e. under gravity with a body force of $\rho g$ N/m$^3$. The bar has a height of $H$, width $2W$ and breadth of $2B$ as indicated, and is made of a material with Young's modulus E and Poisson's ratio $\nu$. The bar is fully clamped at the top. Consider the coordinate system as indicated with the following displacement field, which depends on (x,y,z):
# Displacement in the x-direction: $u_x(x,y,z) = -\frac{\nu\rho g xz }{E}$
# Displacement in the y-direction: $u_y(x,y,z) = -\frac{\nu\rho g yz }{E}$
# Displacement in the z-direction: $u_z(x,y,z) = \frac{\rho g z^2}{2E} + \frac{\nu \rho g (x^2 + y^2)}{2E} - \frac{\rho g H^2}{2E}$

```python
from IPython.display import Image
Image('figures/bar_own_weight.png',embed=True)
```

# 1. Compute the strain matrix (tensor) given the displacement field below: [2] 

```python
import sympy as sp
import numpy as np
rho, g, nu, E, W, B, H, x, y, z = sp.symbols('rho, g, nu, E, W, B, H, x, y, z')

u = -nu*rho*g*x*z/E
v = -nu*rho*g*y*z/E
w = rho*g*z**2/(2*E) + nu*rho*g*(x**2+y**2)/(2*E) - rho*g*H**2/(2*E)

dudx = sp.diff(u,x)
dvdy = sp.diff(v,y)
dwdz = sp.diff(w,z)
dudy = sp.diff(u,y)
dudz = sp.diff(u,z)
dvdx = sp.diff(v,x)
dvdz = sp.diff(v,z)
dwdx = sp.diff(z,x)
dwdy = sp.diff(z,y)

SM_voigt = [dudx,dvdy,dwdz,0.5*(dudy+dvdx),0.5*(dvdz+dwdy),0.5*(dwdx+dudz)]
strain_matrix = sp.Matrix([[SM_voigt[0],SM_voigt[3],SM_voigt[5]],[SM_voigt[3],SM_voigt[1],SM_voigt[4]],[SM_voigt[5],SM_voigt[4],SM_voigt[2]]])
display(strain_matrix)
```

# 2. Compute the stress from the strain matrix computed before: [4]
## The 6x6 material matrix D is already given in symbolic format

```python
D = E/((1 + nu)*(1 - 2*nu))*sp.Matrix([[1-nu,nu,nu,0,0,0],[nu,1-nu,nu,0,0,0],[nu,nu,1-nu,0,0,0],[0,0,0,0.5-nu,0,0],[0,0,0,0,0.5-nu,0],[0,0,0,0,0,0.5-nu]])
D = sp.Matrix(D)
SM_voigt = sp.Matrix(SM_voigt)
Sigma = sp.simplify(sp.Matrix(np.matrix(D)* np.matrix(SM_voigt)))
display(Sigma)
```

# 3. Check whether the computed stress field is in equilibrium: [3]

```python
#fum forces in spesific direction over the lenght of bar.
Sigmafzpos1 = Sigma.subs({'x':0,'y':0,'z': H})[2]
Sigmafzpos2 = Sigma.subs({'x':0,'y':0,'z': 0})[2]
Sigmafzpos3 = Sigma.subs({'x':0,'y':0,'z': H})[4]
Sigmafzpos4 = Sigma.subs({'x':0,'y':0,'z': 0})[4]
Sigmafzpos5 = Sigma.subs({'x':0,'y':0,'z': H})[5]
Sigmafzpos6 = Sigma.subs({'x':0,'y':0,'z': 0})[5]

SumFz = Sigmafzpos1+Sigmafzpos3+Sigmafzpos5

Sigmafxpos1 = Sigma.subs({'x':B,'y':0,'z': 0})[0]
Sigmafxpos2 = Sigma.subs({'x':-B,'y':0,'z': 0})[0]
Sigmafxpos3 = Sigma.subs({'x':B,'y':0,'z': 0})[3]
Sigmafxpos4 = Sigma.subs({'x':-B,'y':0,'z': 0})[3]
Sigmafxpos5 = Sigma.subs({'x':B,'y':0,'z': 0})[5]
Sigmafxpos6 = Sigma.subs({'x':-B,'y':0,'z': 0})[5]

SumFx = Sigmafxpos1+Sigmafxpos3+Sigmafxpos5

Sigmafypos1 = Sigma.subs({'x':0,'y':W,'z': 0})[1]
Sigmafypos2 = Sigma.subs({'x':0,'y':-W,'z': 0})[1]
Sigmafypos3 = Sigma.subs({'x':0,'y':W,'z': 0})[3]
Sigmafypos4 = Sigma.subs({'x':0,'y':-W,'z': 0})[3]
Sigmafypos5 = Sigma.subs({'x':0,'y':W,'z': 0})[4]
Sigmafypos6 = Sigma.subs({'x':0,'y':-W,'z': 0})[4]

SumFy = Sigmafypos1+Sigmafypos3+Sigmafypos5

display(SumFx)
display(SumFy) ## which is equal to the stress due to gravity Hg(rho)
display(SumFz)

print('Which must be equal to:')
display(sp.Matrix([0,0,H*g*rho]))
```

# 3. Check whether the prescribed surface stractions are satisfied: [5]

```python
# All surfaces must be zero traction exept the fixed surface.
display(Sigmafzpos2)#bottom face
display(Sigmafxpos1)#x pos face
display(Sigmafxpos2)#x neg face
display(Sigmafypos1)#y pos face
display(Sigmafypos2)#y neg face
print('All must be equal to 0')
```
