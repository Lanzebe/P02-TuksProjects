## Consider Figure 1 for a beam fixed at the left edge. The beam has length L = 1m, point load of F = 1 kN, and made of steel i.e. E = 210 GPa with solid rectangular cross section of width = 2.5 cm and height 50 mm.

### Force $F = 1000 [N]$
### Length of the beam $L = 1 [m]$
### Young's modulus $E = 210 [GPa]$
### Moment of inertia $ I = 1/12(25/1000)(50/1000)^{3} [m^{4}]  $

## 1) Plot displacement (y [m]), shear force (V [N]) and moment (M [Nm]) along x[m]
### 1.1) Dispalcement plot (y [m]) - 'def' and 'lambda'


```python
import numpy as np
import matplotlib.pyplot as plt

E = 210e9 # [Pa]
L = 1 # [m]
F = 1000 # [N]
I = 1/12*(25/1000)*(50/1000)**3 #(m**4)

# Numerical expression for y

# option 1
def y(x):
    return (F*x**2)/(6*E*I)*(x-3*L)# continous field but there are bounds. e.g. x > 1m  

# option 2
y = lambda x: (F*x**2)/(6*E*I)*(x-3*L) 

dist = np.linspace(0,1,100) # Return evenly spaced numbers over a specified interval. Continuous field!!
disp = y(dist)

plt.figure(1)
plt.plot(dist,disp, color = 'r')
plt.xlabel('Distance, x [m]',Fontsize = 20)
plt.ylabel('Displacement, y [m]',Fontsize = 20)
plt.show()
```

### 1.2) Shear force (V [N]) and Moment (M [Nm]) 

```python
dist = np.linspace(0,1,100)

def V(x):
    return F*np.ones_like(x) #not a function of x -> V = const.

shear = V(dist)
plt.figure(2)
plt.plot(dist,shear)
plt.xlabel('Distance, x [m]',Fontsize = 20)
plt.ylabel('Shear Force, V [N]',Fontsize = 20)

def M(x):
    return F*(x-L)

moment = M(dist)
plt.figure(3)
plt.plot(dist,moment)
plt.xlabel('Distance, x [m]',Fontsize = 20)
plt.ylabel('Moment, M [Nm]',Fontsize = 20)
plt.show()
```

## 2) Find derivatives using 'sympy' : a Python module for symbolic mathematics
### - Calculus tasks such as derivatives, integrals, limits, and series expansions
### 2.1) Create a model for displacement, y using sympy - 'symbols', 'pretty_print'


```python
import sympy as sp
# Initialize symbols
E,L,I,F,x = sp.symbols('E L I F x')

print('Displacement')
y = (F*x**2)/(6*E*I)*(x-3*L)
print('y = {}'.format(y))

print('------------------------------'*2)
print('Pretty print')
sp.pretty_print(y)
```

### 2.2) Find the 1st, 2nd and 3rd derivatives of y using sympy - 'diff', 'simplify'
### - Moment $M = EI\frac{d^{2}y}{dx^{2}}= F(x-L)$
### - Shear force $V = EI\frac{d^{3}y}{dx^{3}} = F$


```python
print('First derivative - Slope')
dydx = sp.diff(y,x)
print('dydx = {}'.format(dydx))

print('------------------------------'*2)
print('Second derivative - Curvature')
d2ydx2 = sp.diff(dydx,x)
print('d2ydx2 = {}'.format(d2ydx2))
M = E*I*d2ydx2
print('M = E*I*d2ydx2 = {}'.format(M))
M = sp.simplify(E*I*d2ydx2)
print('M = E*I*d2ydx2 = {}'.format(M))

print('------------------------------'*2)
print('Third derivative')
d3ydx3 = sp.diff(d2ydx2,x)
print('d3ydx3 = {}'.format(d3ydx3))
V = E*I*d3ydx3
print('V = E*I*d3ydx3 = {}'.format(V))
```

## 3) Calculating numerical values using sympy

### 3.1) Substitute the constants $E,L,I,F$ into the function - subs

```python
# Dictionary
constants = {'F': 1000,
           'L': 1,
           'E': 210e9,
           'I':1/12*(25/1000)*(50/1000)**3}

y = y.subs(constants)
V = V.subs(constants)
M = M.subs(constants)

print('Displacement, y = {}'.format(y))
print('Shear force, V = {}'.format(V))
print('Moment, M = {}'.format(M))
```

### 3.2) Transform sympy expressions to lambda functions - lambdify

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import sympy.plotting as spplt

lam_y = sp.lambdify(x, y,modules = 'numpy') # By default lambdify uses the "math" library
disp = lam_y(dist)

plt.figure(1)
plt.plot(dist,disp)
plt.xlabel('x [m]',Fontsize = 20)
plt.ylabel('Displacement, y [m]',Fontsize = 20)

plt.show()
print(y)
p1 = spplt.plot(y,(x,0,1),ylabel = 'Displacement, y [m]',xlabel = 'x [m]') 
#limited plotting styles and slower than numpy + matplotlib
```

```python
V = V.subs(constants) # V is a constant
shear = V*np.ones_like(dist)

plt.figure(2)
plt.plot(dist,shear)
plt.xlabel('x [m]',Fontsize = 20)
plt.ylabel('Shear Force, V [m]',Fontsize = 20)

M = M.subs(constants) 
print('M = {}'.format(M))
lam_M = sp.lambdify(x,M,modules = 'numpy')
moment = lam_M(dist)

plt.figure(3)
plt.plot(dist,moment)
plt.xlabel('x [m]',Fontsize = 20)
plt.ylabel('Moment, M [Nm]',Fontsize = 20)
```

## 4) Change material properties (Steel 210 GPa -> Aluminium 70GPa) and plot displacement, shear force and moment again.

### 4.1) What changes in displacment do you expect? Plot the displacement of the steel beam and aluminium beam over the length of the beam and verify your answer.


```python
import numpy as np
import matplotlib.pyplot as plt

L = 1 # [m]
F = 1000 # [N]
I = 1/12*(25/1000)*(50/1000)**3 #(m**4)
def y_st(x):
    E = 210e9 # [Pa]
    return (F*x**2)/(6*E*I)*(x-3*L)

def y_al(x):
    E = 70e9 # [Pa]
    return (F*x**2)/(6*E*I)*(x-3*L)


dist = np.linspace(0,1,100) # Return evenly spaced numbers over a specified interval. Continuous field!!
disp_st = y_st(dist)
disp_al = y_al(dist)

plt.figure(1)
plt.plot(dist,disp_st,'r', label = 'steel')
plt.plot(dist,disp_al,'b:', label = 'aluminium')
plt.xlabel('Distance, x [m]',Fontsize = 20)
plt.ylabel('Displacement, y [m]',Fontsize = 20)
plt.legend()
plt.show()
```

### 4.2) What changes in shear force and moment do you expect? Plot the shear force and moment of the steel beam and aluminium beam over the length of the beam and verify your answer.

```python
dist = np.linspace(0,1,100)

def V_al(x):
    return F*np.ones_like(x) #not a function of x -> V = const.

def V_st(x):
    return F*np.ones_like(x) #not a function of x -> V = const.

shear_st = V_st(dist)
shear_al = V_al(dist)

plt.figure(2)
plt.plot(dist,shear_st,'r',label = 'steel')
plt.plot(dist,shear_al,'b:',label = 'aluminium')
plt.xlabel('Distance, x [m]',Fontsize = 20)
plt.ylabel('Shear Force, V [N]',Fontsize = 20)
plt.legend()

def M_st(x):
    return F*(x-L)
def M_al(x):
    return F*(x-L)

moment_st = M_st(dist)
moment_al = M_al(dist)
plt.figure(3)
plt.plot(dist,moment_st,'r',label = 'steel')
plt.plot(dist,moment_al,'b:',label = 'aluminium')
plt.xlabel('Distance, x [m]',Fontsize = 20)
plt.ylabel('Moment, M [Nm]',Fontsize = 20)
plt.legend()
plt.show()
```

# Tips for Superposition

### $\sum F$ Sum of forces
### $\sum M$ Sum of moments
### do not result in enough equations to resolve all variables
### Hence employ principles of superposition


## How to proceed with linear models
### 1) Free body diagram: What loads are present
### 2) How can they be represented using existing linear models, each loading case can use a different model
### 3) Set relationship between displacements: Displacement at a given position = sum of contributions from all different loading conditions to that point
### 4) Set equations = 0, use linear solver to determine unknowns
### 5) Substitute unknowns into governing equations, validate.. and play.
