# Question 4
## Consider Figure 2 for a beam fixed at the left edge. The beam has length L = 1.5m, distributed load of q = 1 kN/m, and made of steel i.e. E = 210 GPa with solid rectangular cross section of width = 2.5 cm and height 50 mm. The beam is supported by a spring that deflects 100mm for every 1 kN of force applied.
## a) Plot the displacement over the length of the beam
## b) Plot the bending moment over the length of the beam
## c) Plot the shear force over the length of the beam

```python
from IPython.display import Image
Image('figures/Tut1_Problem2.png',embed=True)
```

```python
x, v,theta,BM,SF,qp, FA,MA,R = sp.symbols('x, v,theta,BM,SF,qp, FA,MA,R ')
L, q, k, I, E = sp.symbols('L, q, k, I, E')

subies = {'L': 1.5, 'q': 1*10**3 , 'k': (1*10**3)/(0.1), 'I':(0.025*0.05**3)/12, 'E':210*10**9}
#L = 1.5
#q = 1*10**3
#k = (1*10**3)/(0.1)
#I = (0.025*0.05**3)/12
#E = 210*10**9

Eq1 = FA + R - L*(q)
Eq2 = MA + L*R - (L/2)*L*(q)
Eq3 = R*((1/k)+((L**3)/(3*E*I)))-((q*L**4)/(8*E*I))

v= -((q*x**2)/(24*E*I))*(6*L**2-4*L*x+x**2)+ ((3*L-x)*(R*x**2)/(6*E*I))


theta = sp.diff(v,x)
BM = sp.diff(theta,x)*(E*I)
SF = sp.diff(BM,x)
qp = sp.diff(SF,x)

print('v:')
display(sp.simplify(v))
print('theta:')
display(sp.simplify(theta))
print('BM:')
display(sp.simplify(BM))
print('SF:')
display(sp.simplify(SF))
print('q:')
display(sp.simplify(qp))

display(Eq1)
display(Eq2)
display(Eq3)

solution = sp.linsolve([Eq1,Eq2,Eq3],[FA,MA,R]).subs(subies)

R = sp.Matrix(list(solution))[2]
print('R:')
display(R)
```

```python
print('Give Your Solution to Q2a Here')
subies = {'L': 1.5, 'q': 1*10**3 , 'k': (1*10**3)/(0.1), 'I':(0.025*0.05**3)/12, 'E':210*10**9,'R':R}

v = sp.simplify(v.subs(subies))
#display(v)

sp.plotting.plot(v,(x,0,1.5),title='v(x)')
```

## a) Solve for the reaction forces at A, B, D, F and G.

```python
from IPython.display import Image
Image('figures/Tut1_Problem3.png',embed=True)
```

```python
print('Give Your Solution to Q6a Here')

#using superpositions:

L,d,P,q,E,width,height,v,x,I,F_a,F_b,F_d,F_f,F_g,M_a = sp.symbols('L,d,P,q,E,width,height,v,x,I,F_a,F_b,F_d,F_f,F_g,M_a')
d = 0.25

def deflection(F,force_pos,ret_pos):
    if force_pos>=ret_pos:
        return ((F*(ret_pos*d)**2)/(6*E*I))*(-(ret_pos*d)+3*(force_pos*d))
    else:
        return ((F*(force_pos*d)**2)/(6*E*I))*(-(force_pos*d)+3*(ret_pos*d))
    
def deflection_dist(force_pos, ret_pos):
    if force_pos>=ret_pos:
        return (-(q*3.5*d*(ret_pos*d)**2)/(12*E*I))*(3*L+3*(8*d)-2*(ret_pos*d))
    else:
        return (-(q)/(24*E*I))*((ret_pos*d)**4-4*L*(ret_pos*d)**3+6*L**2*(ret_pos*d)**2-4*(ret_pos*d)*(force_pos*d)**3+(force_pos*d)**4)
    
def tot_deflection(pos):
    def1 = deflection(F_b,2,pos)
    def2 = deflection(F_d,4.5,pos)
    def3 = deflection(F_f,8,pos)
    def4 = deflection(F_g,11,pos)
    def5 = deflection(-P,3,pos)
    def6 = deflection_dist(8,pos)
    return sp.simplify(def1+def2+def3+def4+def5+def6)
  
L = 11.5*d
I = (width*height**3)/12
eq1 = sp.simplify(M_a  + 2*d*F_b + 4.5*d*F_d + 8*d*F_f + 11*d*F_g - 3*d*P - (8+(3.5/2))*q*(3.5*d))
eq2 = sp.simplify(F_a + F_b + F_d + F_f + F_g - P - q*(3.5*d))
eq3 = sp.simplify(tot_deflection(2))
eq4 = sp.simplify(tot_deflection(4.5))
eq5 = sp.simplify(tot_deflection(8))
eq6 = sp.simplify(tot_deflection(11))

print('eq1:')
display(eq1)
print('eq2:')
display(eq2)
print('eq3:')
display(eq3)
print('eq4:')
display(eq4)
print('eq5:')
display(eq5)
print('eq6:')
display(eq6)
   
subies = {'d':0.25,'P':1000,'q':1000,'E':210*10**9,'width':0.025,'height':0.05}
solution = sp.linsolve([eq1,eq2,eq3,eq4,eq5,eq6],[F_a,F_b,F_d,F_f,F_g,M_a])

SOLUTION_MATRIX = sp.Matrix(list(solution)) ; display(SOLUTION_MATRIX)
F_a,F_b,F_d,F_f,F_g,M_a = SOLUTION_MATRIX.subs(subies)
print('MA =',M_a)
print('RA =',F_a)
print('RB =',F_b)
print('RD =',F_d)
print('RF =',F_f)
print('RG =',F_g)
```
