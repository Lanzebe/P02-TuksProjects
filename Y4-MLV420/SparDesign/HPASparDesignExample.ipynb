```python

def IntegraF(a,b,Wl):
    k = 1.7857-0.71429
    return ((1.7857*Wl*b) - ((k/(2*6.25))*(b**2)*Wl))-((1.785*Wl*a) - ((k/(2*6.25))*(a**2)*Wl))

def CalcFr(r,Wl):
    Ranges = [0,0.12,0.28,0.44,0.6,0.86,1.12,1.38,1.79,2.2,2.61,3.1625,3.715,4.2675,4.82,5.3725,5.925,6.25]
    if r==1:
        a = Ranges[r-1]
        b = (Ranges[r]+Ranges[r+1])/2
    elif r==16:
        a = (Ranges[r-1]+Ranges[r])/2
        b = Ranges[r+1]
    else:
        a = (Ranges[r-1]+Ranges[r])/2
        b = (Ranges[r]+Ranges[r+1])/2

    force = IntegraF(a,b,Wl)
    #print(force)
    return force

def WingLoading(W):
    return W/(1.25*12.5)
""" 
sum = 0
for i in range(16):
    r = i+1
    val = CalcFr(r,WingLoading(1900))
    print(val)
    sum = sum + val
 """
```
