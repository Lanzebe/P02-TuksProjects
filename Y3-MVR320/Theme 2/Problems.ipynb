```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
k_1,k_1t,l_1,k_2,k_2t,l_2,k_3,l_3,keq1,keq = sp.symbols('k_1,k_1t,l1,k_2,k_2t,l2,k_3,l3,keq1,keq')

k_1t = k_1*l_1
k_2t = k_2*l_2

k_eq1 = (k_1t+k_2t)/l_3

k_eq = sp.simplify((k_eq1*k_3)/(k_eq1+k_3))
k_ans = (((k_1*l_1**2)+(k_2*l_2**2))*(k_3))/(k_3*l_3**2 + k_1*l_1**2 + k_2*l_2**2)
display(sp.simplify(k_eq-k_ans))
```

```python
#Q2.12
l,a,b,E,I,m,k1,k2 = sp.symbols('l,a,b,E,I,m,k1,k2')

c1 = ((3*l*a**2)-(a**3))/(2*(l**3))
c2 = ((3*l*b**2)-(b**3))/(2*(l**3))
kb = (3*E*I)/(l**3)

keq = k1*c1 + k2*c2 + kb

omega_n = (sp.simplify(keq/m))**0.5
omega_n = sp.simplify(omega_n)
display(omega_n)
```

```python
#Q2.127


zeta = 0.1
omega_n = 2
phi = np.pi/2
L = 10
def response(x,zeta, omega_n,C=1):
    return C*np.exp(-zeta*omega_n*x)*np.cos(((1-zeta**2)**0.5)*omega_n*x-phi)
def eeeee(x,zeta, omega_n,C=1):
    return C*np.exp(-zeta*omega_n*x)

x = np.linspace(0,L,1000)
plt.figure()
plt.plot(x,response(x,zeta,omega_n))
plt.plot(x,eeeee(x,zeta,omega_n),'g')
plt.plot(x,-eeeee(x,zeta,omega_n),'g')
plt.plot(x,np.sin(omega_n*x))
plt.plot([0,L],[(1-zeta**2)**0.5,(1-zeta**2)**0.5],'r')
plt.plot([0,L],[-(1-zeta**2)**0.5,-(1-zeta**2)**0.5],'r')
plt.show()
```
